## Numpy Arrays
Last week you learned that lists are good for storing data, but not so good for performing mathematical calculations. Below you will find some examples of creating arrays and using them to perform mathematical calculations. <span style="color:red">Study the code and use print statements to determine what each statement does.
```python
from numpy import array,arange,linspace,sin

a = array([2,3,5]) #creates a list with elements that hold the same number of digits.
print(a)

b = arange(0,10,.1) #creates an array out of a range function.
print(b)

c = linspace(3,8,3) #creates list of evenly spaced numbers. List length will be equal to third parameter value.
print(c)

d = sin(a**2) #calculates sine
print(d)

```

## Plots

Often you will want to plot a function or a data set in python. When doing so you'll want to always keep in mind that computers cannot plot functions. They plot data points and connect them. Below you will find an example of a simple plot. <span style="color:red">After investigating what each line does, add comments to help you remember</span>



```python
from matplotlib import pyplot
from numpy import arange,sin,cos

x = arange(0,5,.1) #assigns x a new value through each iteration: from 0 to 5. It increases by .1 each time.
y = sin(x) #assigns y the value of sine of x in each iteration.
yTwo = cos(x) #assigns yTwo the value of cosine of x in each iteration. 
pyplot.plot(x,y,'r-') #plots the values of x and y on a graph.
pyplot.plot(x,yTwo,'b--') #plots the values of x and yTwo on a graph.
```

## Problem Statement



A skier leaves the ramp of a ski jump with a velocity of $10.0$ m/s, $15.0^\circ$ above the horizontal. The slope is inclined at $50.0^\circ$, and air resistance is negligible. You’d like to know where, on the incline, the skier lands.

## Solution

Below you will find the solution to this problem.  

1. <span style="color:red">Study the code closely and with the help of the book, a classmate, or B. Nelson, figure out what each line is doing.</span>
2. <span style="color:red">Add a comment next to each line describing it.</span>

```python
from numpy import linspace,pi,cos,sin,tan,array
from numpy.linalg import norm
from matplotlib import pyplot
from sys import exit

Nskier = 100
t = linspace(0,3,Nskier) #creats a list of numbers that lead to 3. There are 100 total values in the list.
vi = 10 #initial velocity
theta = 15 * pi/180 #angle of the jump
vix = vi * cos(theta) #velocity of skier on the x axis
viy = vi * sin(theta) #velocity of skier on the y axis
ay = -9.8 #acceleration on the y axis. (gravity)

x = vix * t #finds the position of skier on the x axis as time progresses.
y = viy * t + 1/2 * ay * t**2 #finds the position of skier on the x axis as time progresses.

Nslope = 100 
xslope = linspace(0,30,Nslope)
yslope = -tan(50*pi/180) * xslope

pyplot.plot(xslope,yslope) #creat slope of the incline on which the skier lands.
pyplot.plot(x,y) #prints out position of skier in respect to both, y axis and x axis.
pyplot.show()

#This finds and prints specific point on which the skier lands.
for idx in range(Nskier):
    skierPosition = array([x[idx],y[idx]])
    for index in range(Nslope):
        slopeVec = array([xslope[index],yslope[index]])
        if norm(skierPosition - slopeVec) < 1 and idx > 10:
            print("The landing location is {},{}  m".format(x[idx],y[idx]))
            exit()

#The landing location printed doesn't match the landing location printed on the graph.
```
