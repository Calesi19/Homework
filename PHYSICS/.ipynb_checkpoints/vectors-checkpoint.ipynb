## Loops

A loop is a set of instructions that are to be executed over and over again.  There are two main kinds of loops: 'for' loops and 'while' loops. 'for' loops should be used when you know beforehand exactly how many iterations the loop will execute.  'while' loops should be used when you do not know beforehand how many iterations will be performed, but rather have some criteria that will trigger an exit from the loop.  Below you will find examples and example of a 'for' loop and a 'while' loop. <span style="color:red">After studying the code and output, add an explanation comment (use '# to begin') next to each line.</span>




```python
mysum = 0
for dummy in [3,5,3,2]:
    mysum = mysum + dummy**2
print(mysum)

'''
The loop will go through each item in the list one at a time and will assign the item value to the variable "dummy" through each iteration.
Each time, the number in the list will be squared and then all of them will be added up and stored in the in the variable "mysum."
'''

check = 100
nIter = 0
while check > 0.5:
    check = check /2
    nIter = nIter + 1
print(f"After {nIter} iterations, the value of check dropped to {check}, which was below the threshold of 0.5")

'''
The "check" equals 100; through every iteration that "check" isn't below .5, it will half it, until it is. The variable "nIter" will keep track of how many cycles it went throught by adding 1 to itself in every iteration.
'''
```

## Lists
Often you will want to store more than one data point in a variable. For example, maybe you want to store all of the heights of the students in your class, or their ages, or all of your grades in a class. Instead of assigning each of these numbers to a different variable, we can put them into a single list. Lists are created with square brackets in python ([]), like this:

```python
a = [1,5.2,2.9]  
b = [2,3,4,5,6,7,8,9]
```

Although tempting for the beginner, you cannot use lists to perform mathematical calculations.  To see what I mean, execute the cell below and analyze the results.  Then add comments explaining what happened.


```python
a = [1, 2, 3]
b = [4, 5, 6]
c = a + b
print(c)
d = 2 * a
print(d)
e = a**2
print(e)

'''
In the frist attempt to add both list, it just merged the list into one list with all values, but it didn't add the values.
In the second attempt to multiply the list by 2, it just extended the list and made the values repeat themselves twice, but it didn't multiply the values inside the list.
In the third attempt to square the list, the program failed and returned an error.
'''
```

Although lists are only intended to hold your data (and not for mathematical calculations) there are many functions associated with lists that you will find useful.  Below you will find a few examples.  <span style="color:red">Use print statements to investigate what each function does and add a comment.</span>

```python
a = [1,2,3] #This creates the list.
print(f'{a}')



f = a[0]
g = a[-1]  #This prints the value inside the list with specified index value.
h = a[2]
print(f'\n{f}\n{g}\n{h}')


a.append(4) #This adds a new value to the end of the list.
print(f'\n{a}')



a.pop() #This removes a the last value in the list.
print(f'{a}')


b = len(a) #This returns the amount of values in the list.
print(f'\n{b}')

a.insert(0,5) #This inserts a new item/value in a specified index position.
print(f'\n{a}')

a.sort() #This organizes the list from smallest to largest.
print(f'\n{a}')

b = max(a) #This returns the largest value within the list.
print(f'\n{b}')

c = min(a) #This returns the smallest value within the list.
print(f'\n{c}')

d = sum(a) #This sums all the values inside the list.
print(f'\n{d}')
```

### Problem Statement
[Here](https://drive.google.com/file/d/1FN9rQRp5u8ymTEzdaTdgKEpXpgMoKc-A/view?usp=sharing) you will find a file with some data in it. The file contains a list of 5, two-dimensional vectors, in **component form**.  Calculate the length of each vector and the angle each one makes with respect to the +x axis. 

Note: The data file must be located in the same directory as this file.  


1. Execute the code and observe that the results seem reasonable.
2. Add a comment next to **each line of code**.



```python
import math
import numpy

data = open('sampleData.txt','rb')  

for vec in data:
    theta = math.atan(float(vec[1])/float(vec[0])) * 180/math.pi
    magnitude = math.sqrt(float(vec[0])**2 + float(vec[1])**2)
    print("This vector has length {} and angle {:5.2f} degrees".format(magnitude,theta))

```
